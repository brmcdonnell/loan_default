```python
from sklearn.utils import resample
```

```python
new_df = resample(df, replace = False, # sample without replacement
                    n_samples = 150, # match minority n
                    random_state = 27,
                    stratify=df['status_default']) # reproducible results
```

```python
new_df.drop(columns=[
'status_charged_off',
'status_fully_paid',
'status_in_grace_period',
'status_late__16_30_days_',
'status_late__31_120_days_'], inplace=True)
```

```python
X = new_df.drop(columns=['status_default'])
y = df['status_default']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True,
                                                    random_state=42, stratify=y)
```

```python
rs = RobustScaler() 
X_train_sc = rs.fit_transform(X_train)
X_test_sc = rs.transform(X_test)
```

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
rf = RandomForestClassifier(random_state=42, max_features=0.1,
                                criterion="entropy", class_weight="balanced", verbose=1)
#n_estimators=500,
#

rf.fit(X_train_sc, y_train)
(rf.score(X_train_sc, y_train), rf.score(X_test_sc, y_test))
```

```python
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance',
                                                ascending=False)
```

```python
feature_importances.loc[feature_importances['importance']>0.009]
```

# ADA BOOST 

```python
model = AdaBoostClassifier(random_state=42)
model.fit(X_train,y_train)
y_pred = model.predict(X_test) 
```
